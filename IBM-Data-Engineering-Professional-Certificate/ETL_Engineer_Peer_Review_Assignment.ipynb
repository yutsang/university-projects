## Transform

Using <code>exchange_rate</code> and the `exchange_rates.csv` file find the exchange rate of USD to GBP. Write a transform function that

1. Changes the `Market Cap (US$ Billion)` column from USD to GBP
2. Rounds the Market Cap (US$ Billion)` column to 3 decimal places
3. Rename `Market Cap (US$ Billion)` to `Market Cap (GBP$ Billion)`


```python
def transform(df, exchange_rate):
    # Write your code here
    # convert the 'Market Cap (US$ Billion)' column to a numeric data type
    df['Market Cap (US$ Billion)'] = pd.to_numeric(df['Market Cap (US$ Billion)'], errors='coerce')

    df['Market Cap (US$ Billion)'] = (df['Market Cap (US$ Billion)'] / exchange_rate).round(3)

    # rename the column to 'Market Cap (GBP$ Billion)'
    df = df.rename(columns={'Market Cap (US$ Billion)': 'Market Cap (GBP$ Billion)'})

    return df
```

## Logging Function


Write the logging function <code>log</code> to log your data:


```python
import math

def log(df, column):
    # Write your code here
    # extract the specified column from the dataframe
    data = df[column]

    # compute the natural logarithm of the values in the column
    results = data.apply(math.log)

    # create a new dataframe with the results
    new_df = pd.DataFrame({f"ln({column})": results})

    # return the new dataframe
    return new_df
```
